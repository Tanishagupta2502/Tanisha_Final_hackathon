```python
datam=pd.merge(data,data_labels,on='respondent_id',how='inner')
datam.head()
```

```python
features=datam.drop(columns=['respondent_id','xyz_vaccine','seasonal_vaccine'])
target=datam[['xyz_vaccine','seasonal_vaccine']]
```

```python
object_type=features.select_dtypes(include=['object'])
object_type.head()
```

```python
datam.drop(columns=['respondent_id', 'hhs_geo_region', 'employment_industry', 'employment_occupation'], inplace=True)
```

```python
numerical_features=features.select_dtypes(include=['int64','float64']).columns
numerical_data=features[numerical_features]
numerical_data.head()
```

```python
numerical_imputer=SimpleImputer(strategy='most_frequent')
numerical_data_imputed=numerical_imputer.fit_transform(numerical_data)
numerical_data_df=pd.DataFrame(numerical_data_imputed,columns=numerical_features)
numerical_data_df.head()
```

```python
categorical_features=features.select_dtypes(include=['object']).columns
categorical_data=features[categorical_features]
categorical_data.head()
```

```python
categorical_imputer=SimpleImputer(strategy='most_frequent')
cdi=categorical_imputer.fit_transform(categorical_data)
onehot_encoder=OneHotEncoder(handle_unknown='ignore')
categorical_data_encoded=onehot_encoder.fit_transform(cdi)
```

```python
encoded_df=pd.DataFrame(categorical_data_encoded.toarray(),columns=onehot_encoder.get_feature_names_out(categorical_features))
encoded_df.head()
X=pd.concat([numerical_data_df,encoded_df],axis=1)
X.head()
```

```python
y=target.values
print(np.array(y))
```

```python
X.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import train_test_split
import numpy as np
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.multioutput import MultiOutputClassifier
model=MultiOutputClassifier(LogisticRegression(random_state=42,max_iter=1000))
model.fit(X_train,y_train)
```

```python
from sklearn.metrics import roc_auc_score
```

```python
y_train_pred = model.predict_proba(X_train)
y_train_pred_reshaped = np.hstack([y_train_pred[i][:, 1].reshape(-1, 1) for i in range(len(y_train_pred))])
xyz_auc_train = roc_auc_score(y_train[:, 0], y_train_pred_reshaped[:, 0])
seasonal_auc_train = roc_auc_score(y_train[:, 1], y_train_pred_reshaped[:, 1])
mean_auc_train = (xyz_auc_train + seasonal_auc_train) / 2
print(f'Training Data:')
print(f'XYZ Vaccine AUC: {xyz_auc_train:.4f}')
print(f'Seasonal Vaccine AUC: {seasonal_auc_train:.4f}')
print(f'Mean AUC: {mean_auc_train:.4f}')
```

```python
y_test_pred=model.predict_proba(X_test)
y_test_pred_reshaped = np.hstack([y_test_pred[i][:, 1].reshape(-1, 1) for i in range(len(y_test_pred))])
print(np.array(y_test_pred_reshaped).shape)
print(y_test_pred_reshaped)
```

```python
xyz_auc=roc_auc_score(y_test[:,0],y_test_pred_reshaped[:,0])
seasonal_auc=roc_auc_score(y_test[:,1],y_test_pred_reshaped[:,1])
mean_auc=(xyz_auc+seasonal_auc)/2
```
